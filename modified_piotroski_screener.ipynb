```python
import pandas as pd
from nsepy import get_history
from datetime import date
filename = "mbfm_data - Sheet1.csv"
df = pd.read_csv(filename)

df_tickers = pd.read_csv('ind_nifty50list.csv')

nifty50_tickers = list(df_tickers['Symbol'])
```

```python
def get_historical_price(ticker,start_date,count = 0):
    data = get_history(symbol=ticker, start=start_date, end = start_date)

    if count>=4:
        return 
    try:
        price = float(data['Close'][0])

    except IndexError:
        new_start_date = date.fromordinal(start_date.toordinal()-1)
        count+=1
        price = get_historical_price(ticker,new_start_date,count)
        
    return price
```

```python
def get_shares_outstanding(net_income,ticker, year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'EPS in Rs'):
            eps = float(df[year1][j])
            break
    shares_outstanding = net_income/eps
    return shares_outstanding
```

```python
def get_mv(ticker,year,net_income):
    year1 = 'Mar-'+str(year)
    int_year = 2000+year

    num_shares = get_shares_outstanding(net_income,ticker, year1)
    price = get_historical_price(ticker,date(int_year,3,31))
    if price is None:
        return
    else:
        mv = num_shares * price
    return mv
```

```python
def get_net_income(ticker,year1):
                
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Net Profit'):
            net_inc = float(df[year1][j])
            if(str(net_inc)=='nan'):
                return 
            else:
                return net_inc 
```

```python
def get_ocf(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Cash from Operating Activity +'):
            ocf =float(df[year1][j])
            if(str(ocf)=='nan'):
                return 
            else:
                return ocf
```

```python
def get_se(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Share Capital +'):
            se =float(df[year1][j])
            break
    return se
```

```python
def get_borrowings(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Borrowings'):
            borrowings =float(df[year1][j])
            break
    return borrowings
```

```python
def get_reserves(ticker,year1):
   
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Reserves'):
            reserves =float(df[year1][j])
            break
    return reserves
```

```python
def get_tot_liabilities(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Total Liabilities'):
            tot_liabilities =float(df[year1][j])
            break
    return tot_liabilities
```

```python
def get_current_liabilities(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Other Liabilities +'):
            current_liabilities =float(df[year1][j])
            break
    return current_liabilities
```

```python
def get_other_assets(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Other Assets +'):
            current_assets =float(df[year1][j])
            break
    return current_assets
    
```

```python
def get_sales(ticker,year1):
    
    for j in range(i,len(df.index)):
            
        if (df['Entry'][j] == 'Sales +' or df['Entry'][j] == 'Revenue'):
            sales =float(df[year1][j])
            return sales
```

```python
def get_operating_profit(ticker,year1):
    
    for j in range(i,len(df.index)):
            
        if (df['Entry'][j] == 'Operating Profit' or df['Entry'][j] == 'Financing Profit'):
            op =float(df[year1][j])
            return op
```

```python
def get_eps(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'EPS in Rs'):
            eps = float(df[year1][j])
            return eps
```

```python
def get_pbt(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Profit before tax'):
            pbt = float(df[year1][j])
            return pbt
```

```python
def get_interest(ticker,year1):
   
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Interest'):
            interest = float(df[year1][j])
            return interest
```

```python
def get_dep(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Depreciation'):
            dep = float(df[year1][j])
            return dep
```

```python
def get_tax_rate(ticker,year1):
    
    for j in range(i,len(df.index)):
        
        if (df['Entry'][j] == 'Tax %'):
            tax_rate = float(df[year1][j])
            return tax_rate
```

```python
def calculate_modified_piotroski_score(ticker,year):
    
    global i 
    
    for i in range(0,len(df.index)):
        
        if (df['Ticker'][i] == ticker):
            break

    modified_p_score = 0
    if (year == 11):
        prev_year = 'Mar-09'
    else:
        prev_year = 'Mar-'+str(year-2)
    year0 = 'Mar-'+str(year-1)
    year1 = 'Mar-'+str(year)
    int_year1 = 2000 + int(year)

# Profitability Criteria

    # Positive NI
    
    net_inc = get_net_income(ticker,year1)
    net_inc0 = get_net_income(ticker,year0)
    if net_inc0 is None:
        return (ticker,'Data not available')
    if net_inc is None:
        return (ticker,'Data not available')
    
        #Check condition
   
    if net_inc > 0:
        modified_p_score+=1
    
  
    # Operating CF > NI
    
    ocf = get_ocf(ticker,year1)
    if ocf is None:
        return (ticker,'Data not available')
    
        #Check condition
    elif ocf > net_inc:
        modified_p_score+=1
    
# Leverage, Liquidity, and Source of Funds Criteria

    # Lower ratio of long term debt in current year compared to last year
    
    stock_equity1 = get_se(ticker,year1)
    stock_equity0 = get_se(ticker,year0)
            
    borrowing1 = get_borrowings(ticker,year1)
    borrowing0 = get_borrowings(ticker,year0)
    
    reserves1 = get_reserves(ticker,year1)
    reserves0 = get_reserves(ticker,year0)

    tot_liabilities1 = get_tot_liabilities(ticker,year1)
    tot_liabilities0 = get_tot_liabilities(ticker,year0)

    long_term_debt1 = reserves1 + borrowing1
    long_term_debt0 = reserves0 + borrowing0
    liabilities1 = tot_liabilities1 - stock_equity1
    liabilities0 = tot_liabilities0 - stock_equity0
        
        #Check condition
    if(long_term_debt1/liabilities1 < long_term_debt0/liabilities0):
        modified_p_score+=1
        
    # Higher Current Ratio
    
    current_liabilities1 = get_current_liabilities(ticker,year1)
    current_liabilities0 = get_current_liabilities(ticker,year0)
      
    current_assets1 =get_other_assets(ticker,year1)
    current_assets0 =get_other_assets(ticker,year0)
    
    current_ratio1 = current_assets1/current_liabilities1
    current_ratio0 = current_assets0/current_liabilities0
    
        #Check condition
    if(current_ratio1>current_ratio0):
        modified_p_score+=1
        
    # Net equity issuance
    stock_equity1 = get_se(ticker,year1)
    stock_equity0 = get_se(ticker,year0)
    
        #Check condition
    if (stock_equity1<=stock_equity0):
            modified_p_score+=1

# Operating Efficiency Criteria

    # Higher Gross Margin in current year compared to last year
    
    sales1 =get_sales(ticker,year1)
    sales0 =get_sales(ticker,year0)
    
    operating_profit1 =get_operating_profit(ticker,year1)
    operating_profit0 =get_operating_profit(ticker,year0)
    
        #Check condition
    if (operating_profit0/sales0 > operating_profit1/sales1):
            modified_p_score+=1
    
    # Higher Asset Turnover in current year compared to last year
    
    tot_assets1 =tot_liabilities1
    tot_assets0 =tot_liabilities0

    asset_turnover1 = sales1/tot_assets1
    asset_turnover0 = sales0/tot_assets0
        
        #Check condition
    if (asset_turnover0 > asset_turnover1):
        modified_p_score+=1
        
# Extra Parameters added

    # Price to Book Value < 1
    book_value1 = tot_assets1-liabilities1
    bvps1 = book_value1/get_shares_outstanding(net_inc,ticker, year1)
    price = get_historical_price(ticker,date(int_year1,3,31))
    if price is None:
        return (ticker, 'Price data not available')
    pbv1 = price/bvps1

   
    
        #Check condition
    if (pbv1<1):
        modified_p_score+=1
    

    # Price to Earnings < 10

    eps = get_eps(ticker,year1)
    pe = price/eps
    
    #Check condition
    if pe<=10:
        modified_p_score+=1
        
    # EBITDA to EV value
    ev_value1 = get_mv(ticker,year,net_inc) + borrowing1 - current_assets1
    
    mv0 = get_mv(ticker,year-1,get_net_income(ticker,year0))
    if mv0 is None:
        return (ticker, 'Price data not available')
    else:
        ev_value0 = mv0 + borrowing0 - current_assets0
  
   
    pbt1 = get_pbt(ticker,year1)
    pbt0 = get_pbt(ticker,year0)
    
    interest1 = get_interest(ticker,year1)
    interest0 = get_interest(ticker,year0)
    
    dep1 = get_dep(ticker,year1)
    dep0 = get_dep(ticker,year0)
    
    ebitda1 = pbt1 + interest1 + dep1
    ebitda0 = pbt0 + interest0 + dep1
    
    ev_ratio1 = ev_value1/ebitda1
    
        #Check condition    
    if(ev_ratio1<=10):
        modified_p_score+=1

    # Free Cash Flow to Firm (FCFF)
    ebit1 = pbt1 + interest1
    ebit0 = pbt0 + interest0
    
    wc1= current_assets1 - current_liabilities1
    wc0= current_assets0 - current_liabilities0
    prev_wc = get_other_assets(ticker,prev_year) - get_current_liabilities(ticker,prev_year)
    
    tax_rate1 = get_tax_rate(ticker,year1)
    tax_rate0 = get_tax_rate(ticker,year0)
          
    fcff1 = ebit1*(1-tax_rate1) + dep1 -(wc1-wc0)
    fcff0 = ebit0*(1-tax_rate0) + dep0 -(wc0-prev_wc)
    
        #Check condition
    if (fcff1-fcff0)>0:
        modified_p_score+=1  
    
    return (ticker, modified_p_score)
    
```

```python
for ticker in nifty50_tickers:
    print(calculate_modified_piotroski_score(ticker,11))
```
